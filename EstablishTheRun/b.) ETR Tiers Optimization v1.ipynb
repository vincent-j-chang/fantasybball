```python
## Merge

os.chdir(r"C:\Users\vchang\Documents\Fantasy Bball\DFS\Tiers Analysis")
etr = pd.read_excel(r'C:\Users\vchang\Documents\Fantasy Bball\DFS\ETR_Data\ETR_Daily_' + str(today) + ".xlsx")
dkng = pd.read_csv(r'C:\Users\vchang\Documents\Fantasy Bball\DFS\DKNG_Data\Tiers\DKSalaries_Tiers_' + str(today) + ".csv")


final = pd.merge(etr,
                 dkng[['Name','Roster Position']], 
                 left_on = 'Player',
                 right_on='Name',
                 how='left')

final.dropna(subset=['Roster Position'], inplace=True)

final = final.sort_values(by='DK Points', ascending = False)

final.to_excel("ETR_final_Tiers_"+str(today)+".xlsx", index=False)
os.system("ETR_final_Tiers_"+str(today)+".xlsx")
```
