```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import date
import shutil
import difflib

today = date.today()
today = today.strftime("%m%d20%y")
today
```

```python
bbm = pd.read_csv(r'C:\Users\vchang\Documents\Fantasy Bball\DFS\BBM_Data\BBM_Daily_' + str(today) + ".csv")
dkng = pd.read_csv(r'C:\Users\vchang\Documents\Fantasy Bball\DFS\DKNG_Data\DKSalaries_PlayerPool_' + str(today) + ".csv")
```

```python
bbm['Name'] = bbm['Name'].map(lambda x: difflib.get_close_matches(x, dkng['Name'])[0])
```

```python
bbm
```

```python
final = bbm.set_index('Name').join(dkng.set_index('Name'))
final.reset_index(inplace=True)
```

```python
final.reset_index(inplace=True)
final.to_csv("final"+str(today)+".csv", index=False)
```
