```python
## Merge

os.chdir(r"C:\Users\vchang\Documents\Fantasy Bball\DFS\PlayerPool Analysis")
bbm = pd.read_excel(r'C:\Users\vchang\Documents\Fantasy Bball\DFS\BBM_Data\BBM_Daily_' + str(today) + ".xlsx")
dkng = pd.read_excel(r'C:\Users\vchang\Documents\Fantasy Bball\DFS\DKNG_Data\PlayerPool\DKSalaries_PlayerPool_' + str(today) + ".xlsx")

bbm['nameTeam'] = bbm['Name'] + "_" + bbm['Team'].astype(str)
dkng['nameTeam'] = dkng['Name'] + "_" + dkng['TeamAbbrev'].astype(str)

bbm['Name2'] = bbm['nameTeam'].map(lambda x: difflib.get_closest_match(x, dkng['nameTeam'])[1])

```

```python
final = pd.merge(bbm,dkng[['Name','Salary','Roster Position','nameTeam']], on='nameTeam', how='outer')

```

```python
final['Value'] = final['Expected_Points']/final['Salary']*1000

final = final[final.Value > 0]

final.drop(['nameTeam', 'Name_y'])

final.to_csv("final_PlayerPool_"+str(today)+".csv", index=False)
os.system("final_PlayerPool_"+str(today)+".csv")
```
